```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
import pandas as pd

df = pd.read_csv('home_prices.csv')
print(df)
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price, color = 'red')
plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price)
plt.plot(df.area, model.predict(df[['area']]))
plt.show()
```

```python
model.predict([[4000]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y  = model.intercept_ + model.coef_ * 4000
print(y)
```

```python
df = pd.read_csv('home_prices_input.csv')
print(df)
```

```python
pred = model.predict(df[['area']])
print(pred)
```

```python
df['predicted_prices'] = pred
df.to_csv('home_prices_prediction.csv', index = False)
print(df)
```

```python
df = pd.read_csv('home_prices_prediction.csv')
plt.xlabel('area')
plt.ylabel('prices')
plt.scatter(df.area, df.predicted_prices)
plt.plot(df.area, model.predict(df[['area']]))
plt.show()
```
